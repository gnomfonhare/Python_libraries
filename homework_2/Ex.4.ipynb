```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('fivethirtyeight')
```

```python
creditcard = pd.read_csv('creditcard.csv')
```

```python
class_list = creditcard['Class'].value_counts()
class_list
```

```python
class_list.plot(kind='barh')
plt.show()
```

```python
class_list.plot(kind='barh', logx=True)
plt.show()
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], density=True, bins=20, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], density=True, bins=20, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.xlabel('V1')
plt.show()
```
